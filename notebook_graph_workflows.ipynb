### Translate to script

Sample script to perform PATCH on various policy settings

````powershell
$config = Get-Content ".\config\bw_clientconfiguration.json" -Raw | ConvertFrom-Json
Connect-MgGraph -ClientId $config.ClientId `
    -TenantId $config.TenantId `
    -CertificateThumbprint $config.Thumbprint `
    -NoWelcome
````

````powershell
# 1. GET the current policy
$policyId = "852f466c-bf46-4a33-858f-43aded52b757"
$current = Invoke-MgGraphRequest -Method GET `
    -Uri "https://graph.microsoft.com/beta/deviceManagement/deviceCompliancePolicies/$policyId"

# 2. Inspect before change
Write-Host "Target policy: $($current.displayName)"
Write-Host "Current min password length: $($current.passwordMinimumLength)"
Write-Host "Current active firewall setting: $($current.activeFirewallRequired)"
````

````powershell
# 3. Build PATCH body — include @odata.type, include ONLY fields you are changing
$patchBody = @{
    "@odata.type"            = $current."@odata.type"
    "passwordMinimumLength"  = 12
    "activeFirewallRequired" = $false
}
````

````powershell
# 4. PATCH
Invoke-MgGraphRequest -Method PATCH `
    -Uri "https://graph.microsoft.com/beta/deviceManagement/deviceCompliancePolicies/$policyId" `
    -Body ($patchBody | ConvertTo-Json -Depth 5) `
    -ContentType "application/json"
````

````powershell
# 5. Verify — GET again and confirm the change landed
$updated = Invoke-MgGraphRequest -Method GET `
    -Uri "https://graph.microsoft.com/beta/deviceManagement/deviceCompliancePolicies/$policyId"
Write-Host "New min password length: $($updated.passwordMinimumLength)"
Write-Host "New active firewall setting: $($updated.activeFirewallRequired)"

Disconnect-MgGraph | Out-Null
````

### Understand pagination in responses (odata.nextLink)

```json
# Response structure — always check for nextLink
{
  "@odata.context": "https://graph.microsoft.com/beta/$metadata#...",
  "@odata.nextLink": "https://graph.microsoft.com/beta/deviceManagement/...",  # <-- MORE PAGES
  "value": [ { device1 }, { device2 }, ... ]  # current page only
}
```

### Translate to script

````powershell
$config = Get-Content "C:\vs\config\thiago_clientconfiguration.json" | ConvertFrom-Json
Connect-MgGraph -TenantId $config.TenantId -ClientId $config.ClientId -CertificateThumbprint $config.Thumbprint -NoWelcome

$allDevices = [System.Collections.Generic.List[object]]::new()
$url = "https://graph.microsoft.com/beta/deviceManagement/managedDevices"

do {
    $json = Invoke-MgGraphRequest -Uri $url -Method GET -OutputType Json
    $response = $json | ConvertFrom-Json
    if ($response.value) {
        $allDevices.AddRange($response.value)
    }
    $url = $response.'@odata.nextLink'
}
while ($url)

$allDevices | ConvertTo-Json -depth 30 | Out-File ".\exports\test_filtered.json" -Encoding utf8

Disconnect-MgGraph | Out-Null
````

````powershell
$config = Get-Content "C:\vs\config\thiago_clientconfiguration.json" | ConvertFrom-Json
Connect-MgGraph -TenantId $config.TenantId -ClientId $config.ClientId -CertificateThumbprint $config.Thumbprint -NoWelcome

$allDevices = [System.Collections.Generic.List[object]]::new()
$url = "https://graph.microsoft.com/beta/deviceManagement/managedDevices" +
    "?`$filter=operatingSystem eq 'Windows' and complianceState eq 'noncompliant'" +
    "&`$select=deviceName,userPrincipalName,complianceState,lastSyncDateTime,osVersion" +
    "&`$top=500"

do {
    $json = Invoke-MgGraphRequest -Uri $url -Method GET -OutputType Json
    $response = $json | ConvertFrom-Json
    if ($response.value) {
        $allDevices.AddRange($response.value)
    }
    $url = $response.'@odata.nextLink'
}
while ($url)

$allDevices | ConvertTo-Json -depth 30 | Out-File ".\exports\test_filtered.json" -Encoding utf8

Disconnect-MgGraph | Out-Null
````

### Understanding responses

Copying and Pasting profiles and policies can lead to issues, Graph will return a 400 error. The reason is that the response contains several server-managed read-only fields that are not valid in a POST body.

Properties that MUST be removed before using a GET response as a POST body:

   - id                    — server assigns this on create  
   - createdDateTime       — server managed  
   - lastModifiedDateTime  — server managed  
   - version               — server managed  
   - supportsScopeTags     — informational, not settable  

 @odata.type IS required  
 displayName IS required  
 assignments are separate — they must be posted via /assignments endpoint  

### Translate to script

````powershell
# 159d1544-19ff-4cd8-a02c-10ac2230a4b4
# config
$sourceProfileId   = "2e8435cd-da9f-41ba-a7b7-b3b380d192d8"
$targetDisplayName = "Corp - Windows 10 Baseline (PROD)"
$targetGroupIds    = @(
    "23578512-36fc-411d-a85c-cbae299546b6",  # Group 1
    "29090cfe-29ad-44ff-9619-84d733c908fb"   # Group 2
)
````

````powershell
$config = Get-Content "C:\vs\config\bw_clientconfiguration.json" | ConvertFrom-Json
Connect-MgGraph -TenantId $config.TenantId -ClientId $config.ClientId -CertificateThumbprint $config.Thumbprint -NoWelcome
````

````powershell
# 1. GET the source profile
$uri = "https://graph.microsoft.com/beta/deviceManagement/deviceConfigurations/$sourceProfileId"
$source = Invoke-MgGraphRequest -Method GET -Uri $uri -OutputType Json | ConvertFrom-Json
````

````powershell
# 2. Remove all read-only / server-managed properties
$readOnlyProps = @("id","createdDateTime","lastModifiedDateTime","version","supportsScopeTags")
$postBody = @{}
foreach ($prop in $source.PSObject.Properties) {
    if ($prop.Name -notin $readOnlyProps) { $postBody[$prop.Name] = $prop.Value }
}
# 3. Update display name for target environment
$postBody["displayName"] = $targetDisplayName
````

````powershell
# 4. POST to create the new profile
$newProfile = Invoke-MgGraphRequest -Method POST `
    -Uri "https://graph.microsoft.com/beta/deviceManagement/deviceConfigurations" `
    -Body ($postBody | ConvertTo-Json -Depth 10) `
    -ContentType "application/json"

Write-Host "Created profile ID: $($newProfile.id)"
````

````powershell
# 5. POST assignments as a SEPARATE call
$assignBody = @{
    assignments = @($targetGroupIds | ForEach-Object {
        @{ target = @{ "@odata.type" = "#microsoft.graph.groupAssignmentTarget"; groupId = $_ } }
    })
}

Invoke-MgGraphRequest -Method POST `
    -Uri "https://graph.microsoft.com/beta/deviceManagement/deviceConfigurations/$($newProfile.id)/assign" `
    -Body ($assignBody | ConvertTo-Json -Depth 10) `
    -ContentType "application/json"
````

````powershell
# 6. Confirm assignments — GET and display the assigned groups
$assignmentUri = "https://graph.microsoft.com/beta/deviceManagement/deviceConfigurations/$($newProfile.id)/assignments"
$assignments = Invoke-MgGraphRequest -Method GET -Uri $assignmentUri -OutputType Json | ConvertFrom-Json

foreach ($assignment in $assignments.value) {
    $groupId = $assignment.target.groupId
    $groupUri = "https://graph.microsoft.com/beta/groups/$groupId"
    $group = Invoke-MgGraphRequest -Method GET -Uri $groupUri -OutputType Json | ConvertFrom-Json
    Write-Host "Assigned group: $($group.displayName) ($groupId)"
}
````

````powershell
# Azure Automation Runbook — Event-Driven Restart for Persistently Non-Compliant Devices
# Schedule: Every 4 hours
# Auth: Managed Identity with DeviceManagementManagedDevices.ReadWrite.All
Connect-MgGraph -Identity

$staleThreshold = (Get-Date).ToUniversalTime().AddHours(-4)

# 1. Find Windows devices that are non-compliant
$uri = "https://graph.microsoft.com/beta/deviceManagement/managedDevices" +
    "?`$filter=complianceState eq 'noncompliant' and operatingSystem eq 'Windows'" +
    "&`$select=id,deviceName,complianceState,lastSyncDateTime,managementState,deviceActionResults"

$nonCompliant = [System.Collections.Generic.List[object]]::new()
do {
    $r = Invoke-MgGraphRequest -Method GET -Uri $uri
    $nonCompliant.AddRange($r.value)
    $uri = $r."@odata.nextLink"
} while ($uri)

foreach ($device in $nonCompliant) {
    # 2. Check that the device hasn't already had a reboot action queued
    $pendingReboot = $device.deviceActionResults | Where-Object {
        $_.actionName -eq "rebootNow" -and $_.actionState -in @("pending","active")
    }
    if ($pendingReboot) {
        Write-Output "SKIP: $($device.deviceName) — reboot already queued"
        continue
    }

    # 3. Only act on devices that have been non-compliant long enough
    $lastSync = [datetime]$device.lastSyncDateTime
    if ($lastSync -gt $staleThreshold) {
        Write-Output "SKIP: $($device.deviceName) — non-compliant less than 4 hours"
        continue
    }

    # 4. Issue the reboot action — expect 204 No Content (no response body)
    try {
        Invoke-MgGraphRequest -Method POST \
            -Uri "https://graph.microsoft.com/beta/deviceManagement/managedDevices/$($device.id)/rebootNow"
        Write-Output "RESTART issued: $($device.deviceName)"
    }
    catch {
        Write-Warning "FAILED: $($device.deviceName) — $($_.Exception.Message)"
    }
}

Disconnect-Graph | Out-Null
````

### API Permission assignments for Enterprise Apps

```Application.ReadWrite.All```

````powershell
# 1) Get your service principal (the principal to receive permissions)
$appSp = Get-MgServicePrincipal -Filter "displayName eq 'LACOMMS'"

# 2) Get the Microsoft Graph *resource* service principal (this holds the app roles) ## Graph 00000003-0000-0000-c000-000000000000
$graphResourceSp = Get-MgServicePrincipal -Filter "appId eq '00000003-0000-0000-c000-000000000000'"

# 3) Pick the Graph app role(s) you want to assign (Application permissions)
$role = $graphResourceSp.AppRoles |
Where-Object { $_.Value -eq "ServiceMessage.Read.All" -and $_.AllowedMemberTypes -contains "Application" -and $_.IsEnabled }

# 4) Assign the roles to your app SPN (this is effectively the admin consent)
New-MgServicePrincipalAppRoleAssignment `
    -ServicePrincipalId $appSp.Id `
    -BodyParameter @{
    principalId = $appSp.Id         # who receives the role (your app SP)
    resourceId  = $graphResourceSp.Id    # the resource (Microsoft Graph SP)
    appRoleId   = $role.Id
}
````
